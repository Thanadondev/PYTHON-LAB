```python
import random
from tkinter import *
a = str(random.randint(1000,9999))
print(a)

c=0
life=3
min_v=1000
max_v=9999
allGuess,correct = [],[]
T2 = None
E2 = None

def win():
    win = Tk()
    win.title("Winer")
    Lwin = Label(win, text="ยินดีด้วยคุณชนะ")
    Lwin.pack()
    win.mainloop()
    
def gameover():
    end = Tk()
    end.title("GAME OVER")
    Lend = Label(end, text="พลังชีวิตคุณหมดแล้ว โปรดเริ่มเล่นเกมใหม่อีกครั้ง")
    Lend.pack()
    end.mainloop()

#การทำงานของ Stage1 เมื่อกดปุ่ม
def play_stage1():
    global c, life,allGuess,correct
    T1.delete('1.0', END)
    guess = E1.get().strip()
    g = guess.replace(" ", "")
    count = 0
    if not g.isdigit():
        T1.insert(END, "\nกรุณาใส่ตัวเลขเท่านั้น\n")
    elif len(set(g)) < len(g):
        T1.insert(END, "\nตัวเลขห้ามซ้ำกัน\n")
    elif(len(g)!=3):
        T1.insert(END, "\nใส่เลขได้แค่ 3 หลักเท่านั้น\n")
    elif any(char in num for num in allGuess for char in g):
        T1.insert(END, "\nคุณทายตัวเลขเดิมซ้ำ\n")
    else:
        allGuess.append(g)
        T1.insert(END,"เลขที่คุณทายล่าสุด {0}\n".format(g))
        life-=1
        for j in guess:
            for i in a:
                if(i==j):
                    c+=1
                    count+=1
            
            #เงื่อนไขผ่านด่านแรก
            if c==4:
                T1.insert(END, "\nพบตัวเลขปริศนาทั้งหมดแล้ว\n")
                life+=3
                correct.append(count)
                top.destroy()
                start_stage2()
    #GAMEOVER
    if(life<=0):
        
        T1.insert(END,"เลขที่ถูกต้องคือ : {0}\n".format(a))
        gameover()
    if(count>0):
        correct.append(count)
    T1.insert(END, "พบตัวเลขปริศนา {0} จำนวน, เหลืออีก {1} จำนวน, พลังชีวิตเหลือ {2}\n".format(c, 4 - c, life))
    
#การทำงานของ Stage2 เมื่อกดปุ่ม
def play_stage2():
    global c, life,allGuess,min_v,max_v,correct
    T2.delete('1.0', END)
    guess = E2.get().strip()
    g = guess.replace(" ", "")
    if not g.isdigit():
        T2.insert(END, "\nกรุณาใส่ตัวเลขเท่านั้น\n")
    else:
        GtoNum = int(g)
        #GAMEOVER
        if(life==0):
            T2.delete('1.0', END)
            T2.insert(END,"เลขที่ถูกต้องคือ : {0}\n".format(a))
            gameover()
        #WINNER
        elif(g==a):
            T2.delete('1.0', END)
            T2.insert(END,"เลขที่ถูกต้องคือ : {0}\n".format(a))
            win()
        elif(len(g)!=4):
            T2.insert(END, "\nใส่เลขได้แค่ 4 หลักเท่านั้น\n")
        elif GtoNum<min_v or GtoNum>max_v:
            T2.insert(END, "\nใส่เลขได้แค่ระหว่าง {0}-{1}\n".format(min_v,max_v))
        elif abs(GtoNum - min_v) < abs(GtoNum - max_v):
            closest = GtoNum
            if closest<int(a):
                min_v=(GtoNum+1)
                T2.insert(END,"เลขที่คุณทายคือ : {0} ซึ่งเลขปริศนามีค่ามากกว่านี้\n".format(g))
                life-=1
            else:
                max_v=(GtoNum-1)
                T2.insert(END,"เลขที่คุณทายคือ : {0} ซึ่งเลขปริศนามีค่าน้อยกว่านี้\n".format(g))
                life-=1
        else:
            closest = GtoNum
            if closest>int(a):
                max_v=(GtoNum-1)
                T2.insert(END,"เลขที่คุณทายคือ : {0} ซึ่งเลขปริศนามีค่าน้อยกว่านี้\n".format(g))
                life-=1
            else:
                min_v=(GtoNum+1)
                T2.insert(END,"เลขที่คุณทายคือ : {0} ซึ่งเลขปริศนามีค่ามากกว่านี้\n".format(g))
                life-=1
    
        
        
    T2.insert(END,"เลขปริศนาอยู่ในระหว่าง {0}-{1}\t\t\tพลังชีวิตเหลือ {2}\n".format(min_v,max_v,life))
    T2.insert(END,"เลขที่คุณเคยได้ทายใน Stage1 :")
    
    for ag in allGuess:
        T2.insert(END, f"[{ag}]")
    T2.insert(END,"\n\t\t\t  ")
    for cr in correct:
        T2.insert(END, f"[{cr}]  ")
        
   
def start_stage2():
    global life ,E2, T2,allGuess,min_v,max_v,correct
    bot = Tk()
    bot.title("GAME GUESS NUMBER STAGE 2")
    L2 = Label(bot, text="ใส่ตัวเลข 4 หลัก: ")
    L2.pack()
    E2 = Entry(bot)
    E2.pack()
    B2 = Button(bot, text="ทายด่าน2",command=play_stage2)
    B2.pack()
    T2 = Text(height=30, width=60)
    T2.pack()
    T2.insert(END, f"เกมที่ 2 หาเลขปริศนาให้ครบทุกหลัก\nมีโอกาศในการทาย {life} ครั้ง\nเลขปริศนาอยู่ในระหว่าง 1000-9999 \nเลขที่คุณเคยได้ทายใน Stage1 : ")
    for ag in allGuess:
        T2.insert(END, f"[{ag}]")
    T2.insert(END,"\n\t\t\t  ")
    for cr in correct:
        T2.insert(END, f"[{cr}]  ")
    bot.mainloop()
    
#Start_Stage1   
top = Tk()
top.title("GAME GUESS NUMBER STAGE 1")
L1 = Label(top, text="ใส่ตัวเลข 3 หลัก: ")
L1.pack()
E1 = Entry(top)
E1.pack()
B1 = Button(top, text="ทายด่าน1",command=play_stage1)
B1.pack()
T1 = Text(height=30, width=60)
T1.pack()
top.mainloop()
```
