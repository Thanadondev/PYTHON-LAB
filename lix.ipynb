```python
import tkinter as tk
from tkinter import messagebox, simpledialog

root = tk.Tk()
root.geometry("450x200")

def insert_data():
    name = simpledialog.askstring("Insert Data", "Enter Name:")
    if name:
        phone = simpledialog.askstring("Insert Data", "Enter Phone Number:")
        if phone:
            data = f"{name} - {phone}"
            ins(data)

def edit_data():
    if lb.curselection():
        selected_index = lb.curselection()[0]
        old_data = lb.get(selected_index)
        parts = old_data.split(" - ")
        new_name = simpledialog.askstring("Edit Data", "Edit Name:", initialvalue=parts[0])
        if new_name:
            new_phone = simpledialog.askstring("Edit Data", "Edit Phone Number:", initialvalue=parts[1])
            if new_phone:
                new_data = f"{new_name} - {new_phone}"
                lb.delete(selected_index)
                lb.insert(selected_index, new_data)

def remove_data():
    if lb.curselection():
        selected_index = lb.curselection()[0]
        text = f"Delete\nData: {lb.get(selected_index)}\n"
        ans = messagebox.askquestion("Delete Confirmation", text)
        if ans == "yes":
            lb.delete(selected_index)
        else:
            messagebox.showinfo("Not Deleted", "Changed your mind")

def ins(data):
    lb.insert(tk.END, data)

f1 = tk.Frame(root)
f1.pack(fill=tk.BOTH, expand=1)
tk.Grid.columnconfigure(f1, 0, weight=2)
tk.Grid.columnconfigure(f1, 1, weight=1)
tk.Grid.rowconfigure(f1, 0, weight=1)
lb = tk.Listbox(f1)
lb.grid(row=0, column=0, sticky="nsew")

f2 = tk.Frame(f1)
tk.Grid.columnconfigure(f2, 0, weight=1)
tk.Grid.rowconfigure(f2, (0, 1, 2), weight=1)
f2.grid(row=0, column=1, sticky="news")

b1 = tk.Button(f2, text="INSERT", bg="yellow", command=insert_data)
b1.grid(row=0, column=0, sticky="news")
b2 = tk.Button(f2, text="EDIT", bg="yellow", command=edit_data)
b2.grid(row=1, column=0, sticky="news")
b3 = tk.Button(f2, text="DELETE", bg="yellow", command=remove_data)
b3.grid(row=2, column=0, sticky="news")

root.mainloop()
```
