```python
import typing
from PyQt6 import QtCore, QtWidgets
from PyQt5.QtWidgets import QMainWindow, QPushButton, QTextEdit, QListWidget, QComboBox, QApplication, QMessageBox
from PyQt5 import uic
import sys
class UI(QMainWindow):
    def __init__(self):
        super(UI, self).__init__()
        uic.loadUi("./w6506021411051.ui", self)  # Load UI

        # Button
        self.btn_add = self.findChild(QPushButton, "btn1")
        self.btn_update = self.findChild(QPushButton, "btn2")
        self.btn_del = self.findChild(QPushButton, "btn3")
        self.btn_show = self.findChild(QPushButton, "btn4")

        # text_show 
        self.text_show = self.findChild(QListWidget, "listWidget")

        # textedit
        self.text1 = self.findChild(QTextEdit, "text1")
        self.text2 = self.findChild(QTextEdit, "text2")
        self.text3 = self.findChild(QTextEdit, "text3")

        # combobox
        self.combo = self.findChild(QComboBox, "comboBox")

        # action
        self.btn_add.clicked.connect(self.call_add)
        self.btn_update.clicked.connect(self.call_update)
        self.btn_del.clicked.connect(self.call_del)
        self.btn_show.clicked.connect(self.call_show)
        self.text_show.itemDoubleClicked.connect(self.item_double_clicked)


        # show UI
        self.show()

    def call_add(self):
        if self.text1.toPlainText() == "" or self.combo.currentText() == "" or self.text2.toPlainText() == "" or self.text3.toPlainText() == "":
            m = QMessageBox(self)
            m.setText("Please fill in all fields.")
            m.setWindowTitle("Add Data")
            m.setIcon(QMessageBox.Warning)
            m.show()
        else:
            if self.text2.toPlainText().isdigit():
                text = "Name: " + self.text1.toPlainText() + "\nDepartment: " + self.combo.currentText() + "\nTel No.: " + self.text2.toPlainText() + "\nEmail: " + self.text3.toPlainText()
                self.text_show.insertItem(0,text)
                m = QMessageBox(self)
                m.setText("Data for " + self.text1.toPlainText() + " added successfully.")
                m.setWindowTitle("Add Data")
                m.setIcon(QMessageBox.Information)
                m.show()
            else:
                m = QMessageBox(self)
                m.setText("Tel number must be numeric.")
                m.setWindowTitle("Add Data")
                m.setIcon(QMessageBox.Warning)
                m.show()

    def call_update(self):
        item = self.text_show.currentRow()
        if item >= 0:
            if (
                self.text1.toPlainText() == ""
                or self.combo.currentText() == ""
                or self.text2.toPlainText() == ""
                or self.text3.toPlainText() == ""
            ):
                m = QMessageBox(self)
                m.setText("Please fill in all fields.")
                m.setWindowTitle("Edit Data")
                m.setIcon(QMessageBox.Warning)
                m.show()
        
            else:
                if self.text2.toPlainText().isdigit():
                    new_text = (
                        "Name : "
                        + self.text1.toPlainText()
                        + "\nDepartment : "
                        + self.combo.currentText()
                        + "\nTel No. : "
                        + self.text2.toPlainText()
                        + "\nEmail : "
                        + self.text3.toPlainText()
                    )
                    self.text_show.item(item).setText(new_text)
                    m = QMessageBox(self)
                    m.setText("Update your Data " + self.text1.toPlainText() + " Successfully")
                    m.setWindowTitle("Update")
                    m.setIcon(QMessageBox.Information)
                    m.show()
                else:
                    m = QMessageBox(self)
                    m.setText("Tel number must be numeric.")
                    m.setWindowTitle("Update")
                    m.setIcon(QMessageBox.Warning)
                    m.show()
            
        else:
            m = QMessageBox(self)
            m.setText("No data")
            m.setWindowTitle("Update")
            m.setIcon(QMessageBox.Warning)
            m.show()

    def call_del(self):
        item = self.text_show.currentRow()
        if item >= 0:
            m = QMessageBox.question(self, "Confirm data deletion", "Do you want to delete this information?",
                                    QMessageBox.StandardButton.Yes | QMessageBox.StandardButton.No, QMessageBox.StandardButton.No)
            if m == QMessageBox.StandardButton.Yes:
                self.text_show.takeItem(item)
        else:
            m = QMessageBox(self,text="No data")
            m.setWindowTitle("Delete data")
            m.setIcon(QMessageBox.Icon.Warning)
            m.show()

    def call_show(self):
        self.item = self.text_show.currentRow()
        self.text = self.text_show.currentItem().text()
        print(self.item, self.text)
        self.showmessage()

    def showmessage(self):
        m = QMessageBox(self,text=self.text)
        m.setWindowTitle("Show data")
        m.setIcon(QMessageBox.Icon.Information)
        m.show()
                
    def item_double_clicked(self, item):
        data = item.text()  
        lines = data.split("\n")  

        name = lines[0].replace("Name: ", "")
        department = lines[1].replace("Department: ", "")
        tel_no = lines[2].replace("Tel No.: ", "")
        email = lines[3].replace("Email: ", "")

        self.text1.setPlainText(name)
        self.combo.setCurrentText(department)
        self.text2.setPlainText(tel_no)
        self.text3.setPlainText(email)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = UI()
    window.setWindowTitle("Student Information")
    sys.exit(app.exec_())
```
